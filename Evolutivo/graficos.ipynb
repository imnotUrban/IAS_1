### Gr√°ficos de convergencia

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python

dfBest = pd.read_csv('TXTmin.txt', sep=" ",header=None)
dfProm = pd.read_csv('TXTprom.txt', sep=" ",header=None)
```

```python
fig,ax = plt.subplots(figsize=(22,12))
ax.set_xlabel('numero de iteraciones')
ax.set_ylabel('valor evaluado')
ax.set_title('Gr√°fico de convergencia')

ax.plot(dfBest, label='Mejores')
ax.plot(dfProm, label='Promedio')
ax.legend()

```

```python
print("los m√≠nimos: ")
print(dfBest.tail(5))

print("los promedios: ")
print(dfProm.tail(5))

```
